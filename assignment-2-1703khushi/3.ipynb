```python
import numpy as np
import pandas as pd
import os
from tabulate import tabulate
from random import randrange
import random
from matplotlib import pyplot as plt
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.preprocessing import LabelEncoder
```

# TASK 3.1.1


```python
df = pd.read_csv('./SMAI-Dataset-hc-dataset/new_customers.csv')
label_encoder = LabelEncoder()
df['Genre'] = label_encoder.fit_transform(df['Genre']) # 1 for male, 0 for female
df = df.head(35)
print(df)
```

```python
mean = df.mean()
std = df.std()
df = (df - mean) / std
```

```python
linkage_methods = ['single', 'complete', 'average']
for linkages in linkage_methods:
    Z = linkage(df, method=linkages, metric='euclidean')

    # print(f"Linkage Matrix (Linkage: {linkages})")
    # print(Z)
    dendrogram(Z)
    plt.xlabel('Customer ID')
    plt.ylabel('Distance')
    plt.title(f'Hierarchical Clustering Dendrogram (Linkage: {linkages}))')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()


```

```python
df = df.drop(['CustomerID'], axis=1) # drop customer id

for linkages in linkage_methods:
    Z = linkage(df, method=linkages, metric='euclidean')

    # print(f"Linkage Matrix (Linkage: {linkages})")
    # print(Z)
    dendrogram(Z)
    plt.xlabel('Customer ID')
    plt.ylabel('Distance')
    plt.title(f'Hierarchical Clustering Dendrogram (Linkage: {linkages}))')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

```

The order of the maximum distance for each linkage type is:
1. Complete
2. Average
3. Single

Additionally, the number of clusters and the distances between the clusters appear to have decreased dramatically after eliminating customerID. This is because each customer has a unique identifier, or customerID, making it undesirable to use for clustering.
